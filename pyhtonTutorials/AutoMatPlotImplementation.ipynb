```python
import pandas
auto_data = pandas.read_csv('Auto.csv')
plt.figure(figsize=(8,5))

plt.plot(auto_data.mpg, auto_data.horsepower, 'r.-', label="MPG")
plt.plot(auto_data.mpg, auto_data.acceleration, 'b.-', label="Acceleration")

plt.title("HELLO PLT")
plt.ylabel("MPG-y")
plt.legend()

plt.xticks(auto_data.mpg[::1].tolist()+[40])
plt.show()
```

```python
import pandas
auto_data = pandas.read_csv('Auto.csv')
plt.figure(figsize=(10,10))

plt.plot(auto_data.year, auto_data.mpg, 'r.-', label="MPG")
plt.plot(auto_data.year, auto_data.acceleration, 'b.-', label="Acceleration")

plt.title("HELLO PLT")
plt.ylabel("MPG-y")
plt.legend()

plt.xticks(auto_data.year[::1].tolist())
plt.show()
```

```python
plt.figure(figsize=(4,4), dpi=100)
a = plt.hist(auto_data.mpg, bins=5, color="#6a0dad")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

```python
chevrolet = auto_data.loc[auto_data['name'].str.contains("chevrolet")].count()[0]
plymouth = auto_data.loc[auto_data['name'].str.contains("plymouth")].count()[0]
buick = auto_data.loc[auto_data['name'].str.contains("buick")].count()[0]
amc = auto_data.loc[auto_data['name'].str.contains("amc")].count()[0]
ford = auto_data.loc[auto_data['name'].str.contains("ford")].count()[0]

labels = ['chevrolet', 'plymouth', 'buick', 'amc', 'ford']
colors = ['#6a0dad', '#008000', '#ff5733', '#2e86c1', '#212f3d']
plt.title('3-method')

plt.pie([chevrolet, plymouth, buick, amc, ford], labels=labels, colors=colors)
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)

chevrolet = auto_data.loc[auto_data['name'].str.contains("chevrolet")]['mpg']
ford = auto_data.loc[auto_data['name'].str.contains("ford")]['mpg']
toyota = auto_data.loc[auto_data['name'].str.contains("toyota")]['mpg']

bp = plt.boxplot([chevrolet, ford, toyota], labels=['chevrolet', 'ford', 'toyota'], patch_artist=True)
plt.title('Cars')
plt.ylabel('Overall mpg')

for box in bp['boxes']:
    box.set(color='#6a0dad', linewidth=2)
    box.set(facecolor='#008000')

plt.show()
```

```python
plt.plot(auto_data['name'].str.contains("chevrolet"), auto_data['mpg'], linewidth=1, label='chevrolet', color='#008000')
plt.plot(auto_data['name'].str.contains("ford"), auto_data['mpg'], linewidth=1, label='ford', color='#6a0dad')
plt.legend()
plt.show()
```
