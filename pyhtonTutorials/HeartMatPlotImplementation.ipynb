```python
import pandas
import matplotlib.pyplot as plt

heart_data = pandas.read_csv('heart.csv')
plt.figure(figsize=(20,10), dpi=50)

# chol_data = heart_data.sort_values(by='chol', ascending=False)
# plt.plot(heart_data.age, chol_data, 'r.-', label="Chol")

plt.plot(heart_data.age, heart_data.chol, 'r.-', label="Chol")
plt.plot(heart_data.age, heart_data.thalach, 'b.-', label="Thalach")

plt.title("HELLO-HEART PLT")
plt.ylabel("chol-y && thalach-y")
plt.legend()

plt.xticks(heart_data.age[::1].tolist())
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import magma

graph = figure(title = "Bokeh Scatter Graph")
color = magma(256)
graph.scatter(heart_data['target'], heart_data['thalach'], color=color)
show(graph)
```

```python
graph = figure(title = "Bokeh Bar Chart")
graph.vbar(heart_data['trestbps'], top=heart_data['chol'],
           legend_label = "trestbps VS chol", color='green')
graph.vbar(heart_data['chol'], top=heart_data['thalach'],
           legend_label = "chol VS Size", color='red')
graph.legend.click_policy = "hide"
show(graph)
```
